```python
#Object oriented way of creating plots
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
#add axes method takes a list[] of 4 arguments i.e. left, bottom, width and height argument.
#it shows where the figure will be placed in relation to the blank canvas
```

```python
#in this method we do 3 things in 1 line of code- 
#1. we create a figure using plt.figure(), 
#2. we add_axes by using fig.add_axes() and  
#3. then we plot the axes using the axes.plot() method
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
#plot appearance
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'orange', linewidth=3, linestyle= '--')

#to seach for custom colors one can search for RGB Hex Code on google

#linewidth can control the thickness of the line

#alpha controls the transparency of the line

#linestyle controls the appearance of the line, '--', '.-', ':', 'step' etc
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'purple', linewidth=4, linestyle= '-.')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'red', linewidth=4, linestyle= ':')
```

```python
# linewidth and linestyle can be typed in as lw and ls
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'blue', lw=4, ls= '--')
```

```python
#markers are points marked on the line and one can use various symbols such as 'o', '*', '1', '+' etc
#if you want to make the markers bigger then you can pass the markersize argument
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'blue', lw=3, ls= '-', marker= 'o')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'blue', lw=3, ls= '-', marker= '*', markersize= 10)
```

```python
#to change the color of the marker one can use the markerfacecolor argument, there are also options to change the
#markeredgewidth and markeredgecolor
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'blue', lw=3, ls= '-', marker= 'o', markersize=10, markerfacecolor= 'yellow', markeredgewidth=3,
       markeredgecolor= 'green')
```

```python
#to only show the plot between 0-1
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'blue', lw=3, ls= '--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
